```python
import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40], 'r.', label='circle')
plt.plot([40, 30, 20, 10], 'g^-', label='triangle up')

plt.title('marker')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 20, 30, 40], '-.m*', [1, 2, 3, 4], [50, 60, 70, 80], 'r.', label='circle')
plt.plot([40, 30, 20, 10], 'g^', label='triangle up')

plt.title('marker')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], ls='--', lw=5, c='b', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.title('Style')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([-2, -1, 0, 1, 2, 3], [20, 30, 15, 40, 45, 35], label='선 1')
plt.plot([-2, -1, 0, 1, 2, 3], [25, 10, 35, 20, 15, 25], label='선 2')
plt.title('선 그래프 예제')
plt.xlabel('x축')
plt.ylabel('y축')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

result1 = []
result2 = []
for i in range(13):
    result1.append(random.randint(1, 500))
    result2.append(random.randint(501, 1000))
    
plt.boxplot(result1)
plt.boxplot(result2)
plt.show()
```
